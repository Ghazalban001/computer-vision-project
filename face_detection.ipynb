پردازش ویدیو

```python
#Initialize webcam (0) and load the pre trained Haar Cascade face model
cam = cv2.VideoCapture(0)
face_model = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

#Check if the camera opened correctly if not print error
if not cam.isOpened():
    print("camera didnt open")
else:
    #Start a continuous loop for real time video processing
    while True:
        ret, img = cam.read()
        if not ret:
            print("image capture failed")
            continue

        #Convert frame to grayscale for the detector
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        #Detect faces 
        faces = face_model.detectMultiScale(gray, 1.3, 5)

        #Draw rectangles
        imgout = img.copy()
        if len(faces) > 0:
            for (x, y, w, h) in faces:
                cv2.rectangle(imgout, (x, y), (x + w, y + h), (0, 250, 0), 3)
        else:
            print("face not found")

        #Display the output frame in a window
        cv2.imshow("Face Detection", imgout)
        
        #Press 'q' or 'ESC' (27) to stop the loo
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q') or key == 27:
            print("exit")
            break

    #Release the camera hardware and close all GUI windows
    cam.release()
    cv2.destroyAllWindows()
    
```
