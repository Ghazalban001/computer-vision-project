```python
import cv2
import numpy as np
import pandas as pd
```

```python
cam=cv2.VideoCapture(1)
_,img=cam.read()
cam.release()
```

```python
eye_model=cv2.CascadeClassifier("haarcascade_eye.xml")
face_model=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python

cam = cv2.VideoCapture(0)

while True:
    _, img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face = face_model.detectMultiScale(gray, 1.3, 5)

    if len(face) > 0:
        x, y, w, h = face[0] 
        x2, y2 = x + w, y + h

        gray_face = gray[y:y2, x:x2]
        eye = eye_model.detectMultiScale(gray_face, minSize=(30,30),
                                       scaleFactor=1.1,
                                       minNeighbors=5)
        
        print("x",x,"y",y)
        imageout = img.copy()
        cv2.rectangle(imageout, (x, y), (x2, y2), (0, 250, 0), 3)

        ic = 0
        for (xe, ye, we, he) in eye:
            ic = ic + 1
            cv2.rectangle(imageout, (xe+x, ye+y), (xe+we+x, ye+he+y), (0, 0, 250), 2)
            if ic == 2:
                break
        
        cv2.imshow("ghazal", imageout) 
    
    else:
        cv2.imshow("ghazal", img)
    if cv2.waitKey(1) == ord("q"):
        break

cam.release()
cv2.destroyAllWindows()
```
